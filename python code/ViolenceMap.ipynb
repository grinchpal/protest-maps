```python
import folium
import pandas as pd
import os
import json
from folium import plugins
import geopy
import csv
from geopy.geocoders import Nominatim
```

```python
kk = pd.read_csv('policeviolencecoords.csv')
```

```python
with open('usmap.geojson') as f:
    laArea = json.load(f)
```

```python
laMap = folium.Map(location=[37.6, -95.665], tiles='Stamen Toner', zoom_start=9)
```

```python
folium.GeoJson(laArea).add_to(laMap)
```

```python
for i,row in kk.iterrows():
    folium.CircleMarker((row.latitude,row.longitude), radius=3, weight=2, color='red', fill_color='red', fill_opacity=.5).add_to(laMap)
```

```python
laMap.add_child(plugins.HeatMap(data=kk[['latitude', 'longitude']].to_numpy(), radius=25, blur=10))
```

```python
laMap.save('ViolenceMap.html')
```
