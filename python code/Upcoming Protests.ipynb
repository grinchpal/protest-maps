```python
import geopy
from geopy.geocoders import Nominatim
import pandas as pd
import csv
```

```python
geolocator = Nominatim(user_agent="add-to-coord")
```

```python
df = pd.read_csv('coordspractice.csv')
```

```python
for i,row in df.iterrows():
    location = geolocator.geocode(row.address)
    x = (location.latitude)
    y= (location.longitude)
    with open('protests.csv', 'a') as newFile:
        newFileWriter = csv.writer(newFile)
        newFileWriter.writerow([x, y])
```

```python
with open('usmap.geojson') as f:
    laArea = json.load(f)
```

```python
laMap = folium.Map(location=[37.6, -95.665], tiles='Stamen Toner', zoom_start=9)
```

```python
for lat, lon, website in zip(kk['latitude'], kk['longitude'], kk['website']):
    folium.Marker(
        [lat, lon],
        radius=3,
        weight=2,
        color='red', fill_color='red', fill_opacity=.5,
        popup = website
        ).add_to(laMap)
```
